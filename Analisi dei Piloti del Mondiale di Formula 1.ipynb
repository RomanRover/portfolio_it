#Analisi dei Piloti del Mondiale di Formula 1

Il dataset formula1_data.csv contiene i risultati del mondiale di Formula 1 della Stagione 2008.

Esso contiene 180 righe e le seguenti 5 colonne:

1. Driver: Nome del Pilota
2. Team: Costruttore  per il quale il pilota gareggia
3. Race: Città dove si è svolto il Gran Premio
4. Country: Paese dove si è svolto il Gran Premio
5. Position: Numero compreso tra 0 e 8 che rappresentano l’ordine di arrivo del pilota nella singola gara (0 significa che il pilota non è arrivato tra i primi 8).

Al termine di un gran premio, vengono assegnati i punti ai piloti in base all’ordine di arrivo: 10 al vincitore, 8 al secondo, 6 al terzo, e poi a scalare di 1 punto fino all’ottavo.

#Punto 1

```python
# Define the points system
points_system = {1: 10, 2: 8, 3: 6, 4: 5, 5: 4, 6: 3, 7: 2, 8: 1}
#dizionario con chiavi e corrispettivi valori

# Function to calculate points based on position
def calculate_points(pos):
    return points_system.get(pos, 0)

# Create a new column 'Points' in the DataFrame
df['Points'] = df['Position'].apply(calculate_points)
#la funzione get prende la chiave e applica il suo valore (0 in quanto l'unico per ogni chiave)

# Display the updated DataFrame
df.head()
```

```python
pilot = input('Inserisci il nome del pilota: ')

# Filter the DataFrame for the selected pilot
pilot_df = df[df['Driver'] == pilot]

# Calculate the total points obtained by the selected pilot
total_points = pilot_df['Points'].sum()

# Count the total number of wins for the selected pilot
total_wins = (pilot_df['Position'] == 1).sum()

# Count the total number of podiums for the selected pilot
total_podiums = ((pilot_df['Position'] >= 1) & (pilot_df['Position'] <= 3)).sum()

pilot_records = [f'Punti totali del pilota: {total_points}',f'Vittorie totali: {total_wins}',f'Numero di podi: {total_podiums}']
print(pilot_records)
```

#Punto 2

```python
df.groupby('Driver')['Points'].sum()
```

```python
# Creazione del dizionario vuoto
pilot_points = {}

# Iterazione sui risultati del raggruppamento
for driver, points in df.groupby('Driver')['Points'].sum().items():
    # Aggiunta delle informazioni al dizionario
    pilot_points[driver] = points

# Stampare il dizionario per verificarne il contenuto
print(pilot_points)
```

```python
# Nome del file di output
file_name = "pilot_points.txt"

# Apertura del file in modalità scrittura
with open(file_name, "w") as file:
    # Scrivere l'intestazione per le chiavi
    file.write("Drivers Standings 2008, Formula 1\n")

    # Iterazione sul dizionario e scrittura dei dati nel file
    for driver, points in pilot_points.items():
        file.write(f"{driver},{points}\n")
```
